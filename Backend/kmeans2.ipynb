```python
import numpy as np
import pandas as pd
import sklearn.cluster as cluster
```

```python
Monday = pd.read_csv("data/Weekdays/Monday.csv")
Tuesday = pd.read_csv("data/Weekdays/Tuesday.csv")
Wednesday = pd.read_csv("data/Weekdays/Wednesday.csv")
Thursday = pd.read_csv("data/Weekdays/Thursday.csv")
Friday = pd.read_csv("data/Weekdays/Friday.csv")
Saturday = pd.read_csv("data/Weekdays/Saturday.csv")
Sunday = pd.read_csv("data/Weekdays/Sunday.csv")
```

```python
Mega_df = pd.concat([Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday])
Mega_df.head(9)
```

```python
Night_df = Mega_df.loc[Mega_df['GROUP'] == 'Night']
Late_df = Mega_df.loc[Mega_df['GROUP'] == 'Late Night/ Early Morning']
Morning_df = Mega_df.loc[Mega_df['GROUP'] == 'Morning']
Afternoon_df = Mega_df.loc[Mega_df['GROUP'] == 'Afternoon']
Evening_df = Mega_df.loc[Mega_df['GROUP'] == 'Evening']
```

```python
categories = [('acousticness', 4), ('danceability', 4), ('energy', 5), ('instrumentalness', 4), ('liveness', 3), ('loudness', 4), ('speechiness', 5), ('tempo', 4), ('valence', 4)]
night = {}
for (x, y) in categories:
    temp_df = Night_df[[x]]
    kmeans = cluster.KMeans(n_clusters = y, init = "k-means++")
    kmeans = kmeans.fit(temp_df)
    centers = kmeans.cluster_centers_
    centers = np.reshape(centers, (1, y)).tolist()[0]
    centers.sort()
    night[x] = centers
    
print(night)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import sklearn.cluster as cluster
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
# from sklearn.decomposition import PCA
```

```python
tes_df = pd.read_csv("data/raw2_0.csv")
tes_df.head()
```

```python
test_df = tes_df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']]
test_df.head()
```

```python
scaler = StandardScaler()
scaler.fit(test_df)
```

```python
scaled_data = scaler.transform(test_df)
```

```python
scaled_data.shape
```

```python
clusters_5 = cluster.KMeans(n_clusters = 5)
clusters_5 = clusters_5.fit(scaled_data)
clusters_5.cluster_centers_
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']])
scaled_df.head() 
```

```python
scaled_df['Cluster'] = clusters_5.labels_
scaled_df['date'] = tes_df[['date']].copy()
scaled_df['hour_of_day'] = tes_df[['hour_of_day']].copy()
scaled_df.head()
```

```python
scaled_df.rename(columns={'date': 'dayyyy'}, inplace=True)
scaled_df.head()
# scaled_df_counts = scaled_df.groupby(['dayyyy', 'hour_of_day'])['Cluster'].count()
# scaled_df_counts.head(50)
```
